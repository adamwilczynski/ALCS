```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import statistics

sns.set_style("dark")
```

```python
cooling_rate_percent = [1, 2, 3, 4, 5]
global_energy = [450, 500, 540, 532, 560]

ax = sns.barplot(x=cooling_rate_percent, y=global_energy)
ax.set(ylabel="Global Energy on Finish")
ax.set(xlabel="Cooling Rate [%]")
for i in ax.containers:
    ax.bar_label(i)
```
